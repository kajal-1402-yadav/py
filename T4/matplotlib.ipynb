```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[3,6,9,12]
plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3,4]
y=[3,6,9,12]
plt.plot(x,y,'o',ms=20,mec='r',mfc='w')
plt.show()
plt.plot(x,y,'d',ms=20)
plt.show()
plt.plot(x,y,'*',ms=20)
plt.show()
plt.plot(x,y,'s',ms=20)
plt.show()
plt.plot(x,y,'*-r',ms=20,mfc='r')
plt.show()
plt.plot(x,y,'*:r',ms=20,mfc='r')
plt.show()
plt.plot(x,y,'*--r',ms=20,mfc='r')
plt.show()
plt.plot(x,y,'*-.r',ms=20,mfc='r')
plt.show()
```

```python
x=[1,2,3,4]
y=[21,16,9,12]
plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3,4]
y1=[3,5,9,12]
y2=[2,5,4,5]
plt.plot(x,y1,'o:r')
plt.plot(x,y2,'o--g')
plt.show()
```

```python
x=[1,2,3,4]
y1=[3,5,9,12]
y2=[2,5,4,5]
y3=[12,6,3,7]
plt.plot(x,y1,color='hotpink',linewidth='7')
plt.plot(x,y2,color='magenta',linewidth='5')
plt.plot(x,y3,color='pink',linewidth='7')
plt.xlabel("value1")
plt.ylabel("value2")
plt.title('graph')
plt.show()
```

```python
x=[1,2,3,4]
y1=[3,5,9,12]
y2=[2,5,4,5]
y3=[12,6,3,7]
plt.plot(x,y1,color='hotpink',linewidth='7')
plt.plot(x,y2,color='magenta',linewidth='5')
plt.plot(x,y3,color='pink',linewidth='7')
font1={'family':'serif','color':'red','size':20}
font2={'family':'verdana','color':'green','size':25}
plt.xlabel("x axis",fontdict=font1)
plt.ylabel("y axis",fontdict=font2)
plt.title('graph',fontdict=font2,loc='left')
plt.grid()
plt.legend('1')
plt.show()
```
